# Divergence, Overflow and Python Tuples:

# h(x):
$h(x) = x^5 - 2x^4 + 2$

```python
def h(x):
    return (x**5) - (2 * x**4) + 2
```

# dh(x):
dh(x) = 5 * (x to the 5 - 1 power) - ( (4 * 2) * x to the (4 - 1) power) + 0
<br>
$dh(x) = 5x^4 - 8x^3$

```python
def dh(x):
    return (5 * x**4) - (4 * 2 * x**3)
```

## Define Gradient Descent Function:

```python
################################
# Parameters:
################################
# 1. Derivative function
# 2. Initial guess value
# 3. Learning rate (multiplier)
# 4. Precision
################################
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001, max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]
    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        
        new_x = previous_x - (gradient * multiplier)
        step_size = abs(new_x - previous_x)
        
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        
        if step_size < precision:
            break;
        
    return new_x, x_list, slope_list
```

## Create Data:

```python
import numpy as np
```

```python
x_3 = np.linspace(start=-2.5, stop=2.5, num=1000)
```

```python
print(type(x_3))
```

```python
print(x_3)
```

## Call Gradient Descent Function:

```python
# local_min, list_x, deriv_list = gradient_descent(derivative_func=dh, initial_guess=0.2)
local_min, list_x, deriv_list = gradient_descent(derivative_func=dh, initial_guess=-0.2, max_iter=71)
```
