### Power Rule:
$f(x) = x^n, n <> 0$
<br/>
f'(x) = n * x to the n - 1 power
<p>
Example 1:
<br/>
$f(x) = x^2$
<br/>
f'(x) = 2 * x to the 2 - 1 power   
<br/>
$f'(x) = 2x^1 = 2x$
</p>
<p>
Example 2:
<br/>
g(x) = x^3
<br/>
g'(x) = 3 * x to the 3 - 1 power
<br/>
$g'(x) = 3x^2$    
</p>

<p>
Example 3:
<br/>
h(x) = x to the -100 power
<br/>
h'(x) = -100 * x to the (-100 - 1) power
<br/>
h'(x) = -100 * x to the -101 power
</p>
<p>
Example 4:
<br/>
z(x) = x to the 2.571 power
<br/>
z'(x) = 2.571 * x to the (2.571 - 1) power
<br/>
x'(x) = 2.571 * x to the 1.571 power
</p>
<p>
Example 5:
<br/>
$f(x) = -4x^2 + 7x + 3$
<br/>
f'(x) = (-4 * 2) * (x to the (2 -1) power) + 7 + 0
<br/>
$f'(x) = -8x + 7$
</p>

### g(x):
$g(x) = x^4 - 4x^2 + 5$

```python
def g(x):
    return (x**4) - (4 * x**2) + 5
```

### dg(x):
dg(x) = 4 * (x to the 4 - 1 power) - ( (4 * 2) * x to the (2 - 1) power) + 0
<br>$dg(x) = 4x^3 - 8x$

```python
def dg(x):
    return (4 * x**3) - (8 * x)
```

### Gradient Descent as a Python Function:

```python
################################
# Parameters:
################################
# 1. Derivative function
# 2. Initial guess value
# 3. Learning rate (multiplier)
# 4. Precision
################################
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]
    for n in range(500):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        
        new_x = previous_x - (gradient * multiplier)
        step_size = abs(new_x - previous_x)
        
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        
        if step_size < precision:
            break;
        
    return new_x, x_list, slope_list
```

**Call gradient descent function (passing in all parameters):**

```python
################################
# Required Parameters:
################################
# 1. Derivative function
# 2. Initial guess value
################################
# Optional Parameters:
################################
# 1. Learning rate (multiplier)
# 2. Precision
################################
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=-0.5, 
                                                 multiplier=0.01, precision=0.0001)
print('Local minimum occurs at:', local_min)
print('Number of steps........:', len(list_x))
```

**Call gradient descent function (passing in required parameters only):**

```python
################################
# Required Parameters:
################################
# 1. Derivative function
# 2. Initial guess value
################################
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=-0.1)
print('Local minimum occurs at:', local_min)
print('Number of steps........:', len(list_x))
```
