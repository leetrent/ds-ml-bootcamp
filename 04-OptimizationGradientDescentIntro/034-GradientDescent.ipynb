### Example 1:<br/>(a simple cost function)
$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

**Create Data using numpy.linspace:**

```python
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
type(x_1)
```

```python
print(x_1)
```

### Visualize Data:

```python
plt.xlim([-3, 3])
plt.ylim(0, 8)

plt.xlabel('x', fontsize=16, color='green')
plt.ylabel('f(x)', fontsize=16, color='green')

plt.plot(x_1, f(x_1))
plt.show()
```

### Slope & Derivatives:

Create a python function for the derivative of $f(x)$ called $df(x)$:

```python
def df(x):
    return 2*x + 1
```

**Plot function and derivative side by side:**

```python
# Size the display:
plt.figure(figsize=[15, 5]) # width, height

##############################################
# First Plot on the left: Cost Function
##############################################
plt.subplot(1, 2, 1) # row, column, index

plt.xlim([-3, 3])
plt.ylim(0, 8)

plt.title('Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.grid()

plt.plot(x_1, f(x_1), color='blue', linewidth=3)

##############################################
# Second Plot on the right: Derivative
##############################################
plt.subplot(1, 2, 2) # row, column, index

plt.title('Slope of Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)

plt.grid()

plt.xlim([-2, 3])
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)

plt.show()
```

## Gradient Descent Algorithm:

```python
new_x = 3
#new_x = -3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    step_size = abs(new_x - previous_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if step_size < precision:
        print(f'Loop ran {n} times.')
        break
    
print('Local minimum occurs at:', new_x)
print('Slope or df(x) value...:', df(new_x))
print('Cost or f(x) value......', f(new_x))
```

### Visualize Gradient Descent:

**Superimpose the gradient descent calculations on plot:**

```python
# Size the display:
plt.figure(figsize=[15, 5]) # width, height

##############################################
# First Plot on the left: Cost Function
##############################################
plt.subplot(1, 2, 1) # row, column, index

plt.xlim([-3, 3])
plt.ylim(0, 8)

plt.title('Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.grid()

plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)

x_values = np.array(x_list)
plt.scatter(x_list, f(x_values), color='red', s=100, alpha=0.6)

##############################################
# Second Plot on the right: Derivative
##############################################
plt.subplot(1, 2, 2) # row, column, index

plt.title('Slope of Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)

plt.grid()

plt.xlim([-2, 3])
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

plt.show()
```
