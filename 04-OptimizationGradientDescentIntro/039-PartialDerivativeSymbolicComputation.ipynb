# Data Visualization with 3D Charts

## Minimize function:$$f(x,y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

## Alternative markdown:$$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

### Minimize function expressed in Python:

```python
def f(x, y):
    r = 3**(-x**2 -y**2)
    return 1 / (r + 1)
```

### Create (x, y) data:

```python
import numpy as np
x_4 = np.linspace(start=-2, stop=2, num=200)
y_4 = np.linspace(start=-2, stop=2, num=200)
```

# Visualize the data:

## Generate a 3D Plot:

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
#import color map
from matplotlib import cm
%matplotlib inline
```

```python
fig = plt.figure(figsize=[16, 12])
ax = fig.gca(projection='3d')

ax.set_xlabel('X', fontsize=20)
ax.set_ylabel('Y', fontsize=20)
ax.set_zlabel('f(x,y) - Cost', fontsize=20)

#ax.plot_surface(x_4, y_4, f(x_4, y_4), cmap=cm.coolwarm, alpha=0.4)
ax.plot_surface(x_4, y_4, f(x_4, y_4), cmap=cm.hsv, alpha=0.4)
plt.show()
```

## Partial Derivatives & Symbolic Computation:

```python
from sympy import symbols, diff
```

```python
a, b = symbols('x, y')
```

#### Cost function equation:

```python
f(a, b)
```

#### Cost function equation in Python:

```python
print(f(a,b))
```

#### Partial derivative equation with respect to a:

```python
diff( f(a, b), a)
```

#### Partial derivative equation with respect to a in Python:

```python
print(diff( f(a, b), a))
```

#### Cost function value of f(x, y) when { x: 1.8, y: 1.0 }

```python
f(a, b).evalf(subs={a:1.8, b:1.0})
```

#### Partial derivative value with respect to a when {a: 1.8, b: 1.0}:
