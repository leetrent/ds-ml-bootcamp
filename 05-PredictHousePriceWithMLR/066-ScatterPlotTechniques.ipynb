### Load Data:

```python
from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()
```

### Describe Data:

```python
print(boston_dataset.DESCR)
```

### Prepare Data:

```python
import pandas as pd
```

```python
data = pd.DataFrame(data=boston_dataset.data, columns=boston_dataset.feature_names)
```

#### NOX/DIS Correlation:

```python
nox_dis_corr = data['NOX'].corr(data['DIS'])
```

```python
nox_dis_corr
```

```python
dis_nox_corr = data['DIS'].corr(data['NOX'])
```

```python
dis_nox_corr
```

### Vizualize Data:

#### Using Matplotlib's Scatter Plot:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dotsize = 125

plt.figure(figsize=(12,8))
plt.scatter(x=data['DIS'], y=data['NOX'], alpha=0.6, s=dotsize, color='indigo')
plt.title(f'DIS vs NOX (Correllation = {round(dis_nox_corr, 3)})')
plt.xlabel('Distance from Employment Centers (DIS)')
plt.ylabel('Nitrix Oxide Concentration (NOX)')
plt.show()
```

#### Using Seaborn:

```python
import seaborn as sns
```

```python
sns.set() # resets styling from up above
sns.set_context('talk')
sns.set_style('whitegrid')
sns.jointplot(x=data['DIS'], y=data['NOX'], height=7)
plt.show()
```

```python
sns.set() # resets styling from up above
sns.set_context('talk')
sns.set_style('whitegrid')
sns.jointplot(x=data['DIS'], y=data['NOX'], kind='hex', height=7, color='blue')
plt.show()
```
