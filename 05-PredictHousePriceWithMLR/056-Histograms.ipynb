```python
from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(data=boston_dataset.data, columns=boston_dataset.feature_names)
```

```python
data['PRICE'] = boston_dataset.target
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,6))
nbrOfBars = 50
barColor = '#2196f3'
edgeColor = 'orangered'
plt.hist(data['PRICE'], bins=nbrOfBars, color=barColor, ec=edgeColor)
plt.title('Boston House Prices', fontsize=18)
plt.xlabel('House Prices (thousands $)', fontsize=12)
plt.ylabel('Number of Houses', fontsize=12)
plt.show()
```
