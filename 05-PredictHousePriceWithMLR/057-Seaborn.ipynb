```python
from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(data=boston_dataset.data, columns=boston_dataset.feature_names)
```

```python
data['PRICE'] = boston_dataset.target
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

#### Show Histogram and Probability Density Function:

```python
plt.figure(figsize=(10, 6))
sns.distplot(data['PRICE'], bins=50)
plt.show()
```

#### Show Probability Density Function only (hist=False):

```python
plt.figure(figsize=(10, 6))
sns.distplot(data['PRICE'], bins=50, hist=False)
plt.show()
```

#### Show Histogram only (hist=True, kde=False):

```python
plt.figure(figsize=(10, 6))
sns.distplot(data['PRICE'], bins=50, hist=True, kde=False, color='#fbc02d')
plt.show()
```

### Average Number of Rooms:

#### Using matplotlib:

```python
plt.figure(figsize=(10,6))
barColor = '#00796B'
edgeColor = 'orangered'
plt.hist(data['RM'], color=barColor, ec=edgeColor)
plt.title('Average Number of Rooms', fontsize=18)
plt.xlabel('Number of Rooms', fontsize=12)
plt.ylabel('Number of Houses', fontsize=12)
plt.show()
```

#### Using Seaborn:

```python
plt.figure(figsize=(10, 6))
sns.distplot(data['RM'], bins=50)
plt.show()
```
