```python
from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(data=boston_dataset.data, columns=boston_dataset.feature_names)
```

```python
data['PRICE'] = boston_dataset.target
```

```python
type(data['PRICE'])
```

#### Skew of normal distribution (bell curve) is zero

```python
data['PRICE'].skew()
```

```python
import numpy as np
```

```python
y_log = np.log(data['PRICE'])
```

```python
y_log.head()
```

```python
y_log.tail()
```

```python
type(y_log)
```

```python
data['PRICE'].skew()
```

```python
y_log.skew()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.set_style('whitegrid')
sns.distplot(y_log)
plt.title(f'Log price with skew of {y_log.skew()}')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.distplot(data['PRICE'])
plt.title(f"Log price with skew of {data['PRICE'].skew()}")
plt.show()
```

```python
sns.lmplot(x='LSTAT', y='PRICE', data=data, height=7,
          scatter_kws={'alpha': 0.6}, line_kws={'color': 'darkred'})
plt.show()
```
